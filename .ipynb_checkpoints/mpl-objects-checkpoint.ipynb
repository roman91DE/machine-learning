```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
```

```python
import random
```

```python
%matplotlib notebook
```

```python
fig, ax = plt.subplots()

x = np.array([random.random() for _ in range(100)])

ax.pie(x)
fig.show()
```

```python
def split(x):
    return np.where(x > .5, 1, 0)

fig, ax = plt.subplots()
y = split(x)

ax.plot(x,y)
fig.show()
```
