```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
a = np.array([0,4])
b = np.array([6,0])
c = a + b
```

```python
def plot_vec(ax, name, vec, orig=np.array([0,0])):
    """draw vector(from origin) and annotate name to ax"""
    ax.annotate(f"{name}({vec[0]},{vec[1]})", vec, fontsize=24)
    ax.quiver(
        orig[0],
        orig[1],
        vec[0],
        vec[1],
        angles="xy",
        scale_units="xy",
        scale=1
    )
```

```python
fig, ax = plt.subplots(
    figsize=[15, 12]
)
ax.set_xlabel("X", fontsize=30)
ax.set_ylabel("Y", fontsize=30)
ax.set_xlim(-5,10)
ax.set_ylim(-5,10)
ax.grid()


plot_vec(ax,"a", a)
plot_vec(ax, "b", b)
plot_vec(ax, "a+b", a+b)
```
