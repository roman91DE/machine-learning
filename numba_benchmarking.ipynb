```python
import numpy as np
from numba import njit
import math as ma
from time import time
```

```python
X = np.arange(-500,500,.001, dtype=np.float64)
X.size
```

```python
def e(x):
  return [ma.exp(n) for n in x]

def e_vec(X):
  return np.exp(X)

e_jit = njit(e)

e_vec_jit = njit(e_vec)
```

```python
def timer(f, *args):
  t0 = time()
  f(*args)
  tn = time()
  print(f'Time elapsed: {tn - t0:.7f}')
```

```python
funcs = {
  'for loop' : e,
  'vectorized': e_vec,
  'compiled for loop' : e_jit,
  'compiled & vectorized': e_vec_jit
}
```

```python
for id, f in funcs.items():
  print(f'Method: {id}')
  print('First Run:')
  timer(f, X)
  print('Second Run:')
  timer(f, X)
  print()
```
