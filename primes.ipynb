```python
import numpy as np
import math
from matplotlib import pyplot as plt
```

```python
def isPrime(n):
    if n < 2:
        return False
    for div in range(2, math.floor(math.sqrt(n)+1), 1):
        if n % div == 0:
            return False
    return True

# np_isPrime = np.vectorize(_isPrime)
```

```python
primes = np.array([n for n in range(1000) if isPrime(n)])
X = np.arange(primes.size)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(28,12)
ax.scatter(X,primes)
ax.set_title("Prime Numbers", fontsize=34)
ax.grid()
```

```python
def get_Twin(n):
    if isPrime(n+2):
        return (n, n+2)
    else:
        return False

twlist = []

for n in primes:
    tmp = get_Twin(n)
    if tmp:
        twlist.append(tmp)

twins = np.array(twlist)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(28,12)
X = np.arange(0, twins[:,0].size)
ax.scatter(X, twins[:,0], color="blue")
ax.scatter(X, twins[:,1], color="yellow")

ax.set_title("Prime Twins", fontsize=34)
ax.grid()
```

```python
functs = {
    "f(x) = x^2"    : lambda x: x ** 4,
    "f(x) = x^(1/2)": lambda x: x ** (1/3),
    
    "f(x) = e^x":     lambda x: np.exp(2*x),
    "f(x) = e^(-x)":  lambda x: np.exp(-42*x),
    
    "f(x) = sin(x)":  lambda x: np.sin(x**2),
    "f(x) = cos(-x)": lambda x: np.cos(-x*1331),
}
```

```python
def plotIt(X, functs):
    
    ncols = math.ceil(len(functs) / 2)
    fig, axs = plt.subplots(nrows=2, ncols=ncols)
    fig.set_size_inches(42,22)
    
    keys = list(functs.keys())
    idx = 0
    
    for row in axs:
        for ax in row:
            
            
            if idx == len(keys):  # handle uneven number of functions
                return
            
            ax.plot(X, functs[keys[idx]](X))
            ax.set_title(keys[idx], size=24)
            idx += 1
            
            # style
            ax.set_xlabel("x", size=16)
            ax.set_ylabel("f(x)", )
            ax.grid()
            
            # scale
            ax.set_xlim(X[0],X[-1])
            ax.set_ylim(X[0],X[-1])
        
            
plotIt(np.arange(-50,100), functs)
```
