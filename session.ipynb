```python
import numpy as np
import math
```

```python
nats = np.arange(10000)
```

```python
nats
```

```python
def isPrime(n):
    if n < 2:
        return False
    for div in range(2, math.floor(math.sqrt(n)+1), 1):
        if n % div == 0:
            return False
    return True

isPrime_vectorized = np.vectorize(isPrime)
```

```python
%timeit res1 = isPrime_vec(nats)
```

```python
def slow():
    res2 = [False for _ in range(len(nats))]
    for val in nats:
        if isPrime(val):
            res2.append(True)
        else:
            res2.append(False)  

%timeit slow()
```
