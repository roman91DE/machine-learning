```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
a = np.array([0,4])
b = np.array([6,0])
```

```python
def plot_vec(ax, name, vec, orig=np.array([0,0])):
    """
    draw vector and annotate its name to matplotlib.axes.Axes
    """
    ax.annotate(f"{name}({vec[0]},{vec[1]})", vec, fontsize=24)
    ax.quiver(
        orig[0],
        orig[1],
        vec[0],
        vec[1],
        angles="xy",
        scale_units="xy",
        scale=1
    )
```

# Addition

```python
fig, ax = plt.subplots(
    figsize=[15, 12]
)
ax.set_xlabel("X", fontsize=30)
ax.set_ylabel("Y", fontsize=30)
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
ax.grid()


plot_vec(ax,"a", a)
plot_vec(ax, "b", b)
plot_vec(ax, "a+b", a+b)
```

# Substraction

```python
fig, ax = plt.subplots(
    figsize=[15, 12]
)
ax.set_xlabel("X", fontsize=30)
ax.set_ylabel("Y", fontsize=30)
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
ax.grid()


plot_vec(ax,"a", a)
plot_vec(ax, "b", b)
plot_vec(ax, "a-b", a-b)
plot_vec(ax, "b-a", b-a)
```

# Multiplication

```python
fig, ax = plt.subplots(
    figsize=[15, 12]
)
ax.set_xlabel("X", fontsize=30)
ax.set_ylabel("Y", fontsize=30)
ax.set_xlim(-10,20)
ax.set_ylim(-10,20)
ax.grid()

d = np.array([4,2])
e = np.array([-2,7])

plot_vec(ax,"d", d)
plot_vec(ax, "e", e)
plot_vec(ax, "d*e", d*e)
```
