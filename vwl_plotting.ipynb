## Linear model 

### Linear function
> $$ p=m * x + b $$

### Slope
> $$ m = \delta {p} - \delta {x} = \frac {(p_2 - p_1)} {(x_2 - x_1)} $$

```python
def p(x, m, b):
    return (m * x) + b

def m(x1, x2, p1, p2):
    return ((p2 - p1) / (x2 - x1))
```

```python
# model parameters
env = {
    "p_min": 0,
    "p_max": 1000,
    "x_min": 0,
    "x_max": 1000
    }


# functions: f(quantity) -> price
def supply(x):
    return p(
        x,
        m(env["x_min"], env["x_max"], env["p_min"], env["p_max"]),
        env["p_min"]
    )

def demand(x):
    return p (
        x,
        m(env["x_min"], env["x_max"], env["p_max"], env["p_min"]),
        env["p_max"]
    )
    
# plotting ranges
prices = np.linspace(
    env["p_min"], env["p_max"], (env["p_max"] - env["p_min"])
    )
x = np.linspace(
    env["x_min"], env["x_max"], (env["x_max"] - env["x_min"])
    )
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for axe in axes:
    axe.set_xlabel("Menge [Stück]")
    axe.set_ylabel("Preis [EUR]")

axes[0].plot(prices, supply(prices), label="Angebot")
axes[1].plot(prices, demand(prices), label="Nachfrage")

axes[0].legend()
axes[1].legend()
```

```python
fig, ax = plt.subplots()

ax.set_xlabel("Menge [Stück]")
ax.set_ylabel("Preis [EUR]")

ax.plot(prices, supply(prices), label="Angebot")
ax.plot(prices, demand(prices), label="Nachfrage")

ax.legend()
```

## Equilibrium

```python
fig, ax = plt.subplots()

ax.plot(prices, supply(prices), label="Angebot")
ax.plot(prices, demand(prices), label="Nachfrage")

pmean = np.full(len(prices), np.mean(prices))
xmean = np.full(len(x), np.mean(x))

ax.plot(prices, xmean)
ax.plot(pmean, x)

ax.legend()
```
