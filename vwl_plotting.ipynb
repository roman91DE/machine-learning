```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
```

# Supply and Demand

## Linear Function
$$ y=m * x + b $$
$$ m = \delta {y} - \delta {x} = \frac {(y_2 - y_1)} {(x_2 - x_1)} $$

```python
# linear functions

def f(x, m, b):
    return (m * x) + b

def m(x1, x2, p1, p2):
    return ((p2 - p1) / (x2 - x1))
```

```python
# model parameters
p_min, p_max = 0, 1000
x_min, x_max = 0, 1000

# functions: f(quantity) -> price
def supply(x):
    return f(
        x,
        m(x_min, x_max, p_min, p_max),
        p_min
    )

def demand(x):
    return f(
        x,
        m(x_min, x_max, p_max, p_min),
        p_max
    )
    
# plotting ranges
prices = np.linspace(p_min, p_max, (p_max - p_min))
x = np.linspace(x_min, x_max, (x_max - x_min))
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for axe in axes:
    axe.set_xlabel("Menge [Stück]")
    axe.set_ylabel("Preis [EUR]")

axes[0].plot(prices, supply(prices), label="Angebot")
axes[1].plot(prices, demand(prices), label="Nachfrage")

axes[0].legend()
axes[1].legend()
```

```python
fig, ax = plt.subplots()

ax.set_xlabel("Menge [Stück]")
ax.set_ylabel("Preis [EUR]")

ax.plot(prices, supply(prices), label="Angebot")
ax.plot(prices, demand(prices), label="Nachfrage")

ax.legend()
```

## Equilibrium

```python
fig, ax = plt.subplots()

ax.plot(prices, supply(prices), label="Angebot")
ax.plot(prices, demand(prices), label="Nachfrage")

pmean = np.full(len(prices), np.mean(prices))
xmean = np.full(len(x), np.mean(x))

ax.plot(prices, xmean)
ax.plot(pmean, x)

ax.legend()
```
