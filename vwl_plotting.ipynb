```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
```

# Microeconomics 
1. Basic Supply & Demand

```python
# linear functions

def p(x, m, b):
    return (m * x) + b

def m(x1, x2, p1, p2):
    return ((p2 - p1) / (x2 - x1))
```

```python
# model parameters
p_min, p_max = 0, 100
x_min, x_max = 0, 100

# functions: f(quantity) -> price
def supply(x):
    return p(
        x,
        m(x_min, x_max, p_min, p_max),
        p_min
    )

def demand(x):
    return p(
        x,
        m(x_min, x_max, p_max, p_min),
        p_max
    )
    
# plotting ranges
prices = np.linspace(p_min, p_max, (p_max - p_min))
x = np.linspace(x_min, x_max, (x_max - x_min))
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for axe in axes:
    axe.set_xlabel("Menge [Stück]")
    axe.set_ylabel("Preis [EUR]")

axes[0].plot(prices, supply(prices), label="Angebot")
axes[1].plot(prices, demand(prices), label="Nachfrage")

axes[0].legend()
axes[1].legend()
```
